```python
#1. -Tampilkan image dalam channel Red-Green !
img_rg = img.copy()
img_rg[:,:,2] = 0
plt.imshow(img_rg)
```

```python
# -Tampilkan image dalam channel Green-Blue
img_gb = img
img_gb[:,:,0] = 0
plt.imshow(img_gb)
```

```python
#2. Tampilkan image baris ke 10-100, kolom 10-100!
img_k = img[10:100,10:100]
plt.imshow(img_k)
```

```python
#3. Tampilkan image baris ke 5-30, semua kolom, channel Green saja!
plt.imshow(img[5:30])
```

```python
#4. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image. disarankan menggunakan bilangan acak/random!
imageWithRectangle = cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig = plt.figure(figsize=(10,10))
cv.rectangle(imageWithRectangle, pt1=(500,800), pt2=(400,200), color=(0,255, 255), thickness = 10)
cv.rectangle(imageWithRectangle, pt1=(400,200), pt2=(200,500), color=(0,255,0), thickness = 10)
cv.rectangle(imageWithRectangle, pt1=(200,500), pt2=(700,400), color=(255,255,0), thickness = 10)
cv.rectangle(imageWithRectangle, pt1=(400,500), pt2=(200,700), color=(255,0,0), thickness = 15)
cv.rectangle(imageWithRectangle, pt1=(500,100), pt2=(200,700), color=(0,0,255),thickness=10)
plt.imshow(imageWithRectangle)
```

```python
#5. Tampilkan image dengan posisi terbalik!
img_t = cv.flip(cv.cvtColor(img,cv.COLOR_BGR2RGB),0)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(111)
ax.imshow(img_t)
```

```python
#6. Buat rectangle dan circle pada bagian wajah dari image kucingku!
img = cv.imread('/content/drive/MyDrive/PCVK_MM/Images/kitten01.jpg')
fig = plt.figure(figsize=(5,5))

ax = fig.add_subplot(111)
cv.rectangle(img, pt1=(300,100), pt2=(600,400), color=(255, 0, 30), thickness = 20)
cv.circle(img, center=(500, 200), radius=150, color=(180, 105, 255), thickness = 20)
ax.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```
